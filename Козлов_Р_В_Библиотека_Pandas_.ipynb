```python
#Импорт библиотек
import pandas as pd
import numpy as np
```

```python
#Задаем переменную
df = pd.read_csv('https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/telecom_churn.csv')
#Выводим первые 20 строк
df.head(20)
```

```python
#Прописываем опции для отображения в Pandas большего количества строк и столбцов
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
#Смотрим количество строк и столбцов в датасете
df.shape
```

```python
#Смторим названия колонок
df.columns
```

```python
#Меняем тип данных в столбике Churn
df['Churn']=df['Churn'].astype('int64')

#Выводим основные статичтические показатели датасета
df.describe()
```

```python
#описательная статистика для типов данных object и bool
df.describe(include=['object','bool'])
```

```python
#Кол-во значений 0 и 1 по параметру Churn
df['Churn'].value_counts()
```

```python
#Распределение по параметру Area code в долях
df['Area code'].value_counts(normalize=True)
```

```python
#Первые 5 строк по убыванию параметра Total day charge
df.sort_values(by='Total day charge',ascending=False).head()
```

```python
#Сортировка по двум параметрам
df.sort_values(by=['Churn','Total day charge'],ascending=[True,False]).head()
```

```python
#Средняя по параметру Churn
df['Churn'].mean()
```

```python
#Средние по всем параметрам,где Churn = 1
df[df['Churn']==1].mean()
```

```python
#Сколько в среднем в течение дня разговаривают по телефону нелояльные пользователи
df[df['Churn']==1]['Total day minutes'].mean()
```

```python
#Максимальная длина международных звонков среди лояльных пользователей, не пользующихся услугой международного роуминга 
df[(df['Churn']==0)& (df['International plan']=='No')]['Total intl minutes'].max()
```

```python
#Выборка датасета по индексу методом loc
df.loc[0:5,'State':'Area code']
```

```python
#Выборка датасета по индексу методом iloc
df.iloc[0:5,0:3]
```

```python
#Последняя строка датасета
df[-1:]
```

```python
#Применение функции max к каждому столбцу датасета при помощи библиотеки Numpy
df.apply(np.max)
```

```python
#Меняем значение ячееек функцией map
d= {"No" : False,"Yes" :True}
df['International plan']=df['International plan'].map(d)
df.head()
```

```python
#Меняем значение ячееек функцией replace
df=df.replace({'Voice mail plan':d})
df.head()
```

```python
#Описательная статистика по сгруппированному значению параметра Churn
columns_to_show=['Total day minutes', 'Total eve minutes', 'Total night minutes']
df.groupby(['Churn'])[columns_to_show].describe()
```

```python
#Описательная статистика по сгруппированному значению параметра Churn через функцию agg и библиотекой Numpy
columns_to_show=['Total day minutes', 'Total eve minutes', 'Total night minutes']
df.groupby(['Churn'])[columns_to_show].agg([np.mean,np.max,np.std,np.min])
```

```python
#Таблица сопряженностей для показателей International plan и Churn
pd.crosstab(df['Churn'],df['International plan'])
```

```python
#Таблица сопряженностей для показателей International plan и Churn, относительные показатели
pd.crosstab(df['Churn'],df['Voice mail plan'],normalize = True)
```

```python
#Сводная таблица
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'],['Area code'], aggfunc='mean').head(10)
```

```python
#Добавляем столбик Total calls 
total_calls=df['Total day calls']+df['Total eve calls']+df['Total night calls']+df['Total intl calls']
df.insert(loc=len(df.columns), column = 'Total calls', value = total_calls)
```

```python
df.head()
```

```python
#Добавляем столбик Total charge
df['Total charge']= df['Total day charge']+df['Total eve charge']+df['Total intl charge']+df['Total night charge']
df.head()
```

```python
#Удаляем столбики Total calls, Total charge
df=df.drop(['Total calls', 'Total charge'], axis =1)
df.head()
```

```python
#Удаление строки по индексу
df.drop([1,2]).head()
```

```python
#Перекрестная таблица параметров Churn, International plan с общими итогами
pd.crosstab(df['Churn'], df['International plan'], margins=True)
```

```python
#Перекрестная таблица параметров Churn, Customer service calls с общими итогами
pd.crosstab(df['Churn'], df['Customer service calls'], margins=True)
```

```python
df['Many_service_calls']=(df['Customer service calls'] > 3).astype('int')
pd.crosstab(df['Many_service_calls'], df['Churn'], margins=True)
```

```python
pd.crosstab(df['Many_service_calls'] & df['International plan'], df['Churn'])
```
